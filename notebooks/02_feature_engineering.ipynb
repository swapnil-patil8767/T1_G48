```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0.05)
```

```python


columns = x.columns[sel.get_support()]
columns
```

```python
corr_matrix = x.corr()
```

```python

columns = corr_matrix.columns


columns_to_drop = []


for i in range(len(columns)):
    for j in range(i + 1, len(columns)):

        if corr_matrix.loc[columns[i], columns[j]] > 0.95:
            columns_to_drop.append(columns[j])

print(len(columns_to_drop))
```

```python
columns_to_drop = set(columns_to_drop)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

y_test
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
y_test
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x_train, y_train)
```

```python

importances = model.feature_importances_
```

```python


feature_ranking = pd.DataFrame({
    'Feature': x_train.columns,
    'Importance': importances
})


feature_ranking = feature_ranking.sort_values(by='Importance', ascending=False)


feature_ranking = feature_ranking.reset_index(drop=True)


print(feature_ranking)
```

```python

top_45_features = feature_ranking.head(45)['Feature'].tolist()


x_top_45 = x[top_45_features]


display(x_top_45.head())
```

```python
df
```

```python

df_final_training = df[['label'] + top_45_features]


display(df_final_training.head())
```
