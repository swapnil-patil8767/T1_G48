```python
def create_time_windows(start_time, end_time, window_size, overlap):
    """
    Create overlapping time windows
    """
    windows = []
    current_start = start_time

    while current_start + window_size <= end_time:
        window_end = current_start + window_size
        windows.append((current_start, window_end))
        current_start += (window_size - overlap)

    return windows

```

```python
def aggregate_window_data(eeg_data, facial_data, window_start, window_end):
    """
    Aggregate all sensor data within a time window
    """

    eeg_window = eeg_data[(eeg_data['TimeStamp'] >= window_start) &
                         (eeg_data['TimeStamp'] < window_end)]
    facial_window = facial_data[(facial_data['Timestamp'] >= window_start) &
                               (facial_data['Timestamp'] < window_end)]


    eeg_features = {}
    eeg_cols = ['Delta_TP9', 'Delta_AF7', 'Delta_AF8', 'Delta_TP10',
       'Theta_TP9', 'Theta_AF7', 'Theta_AF8', 'Theta_TP10', 'Alpha_TP9',
       'Alpha_AF7', 'Alpha_AF8', 'Alpha_TP10', 'Beta_TP9', 'Beta_AF7',
       'Beta_AF8', 'Beta_TP10', 'Gamma_TP9', 'Gamma_AF7', 'Gamma_AF8',
       'Gamma_TP10', 'RAW_TP9', 'RAW_AF7', 'RAW_AF8', 'RAW_TP10', 'AUX_RIGHT',
       'Accelerometer_X', 'Accelerometer_Y', 'Accelerometer_Z', 'Gyro_X',
       'Gyro_Y', 'Gyro_Z', 'HeadBandOn', 'HSI_TP9', 'HSI_AF7', 'HSI_AF8',
       'HSI_TP10', 'Battery']

    for col in eeg_cols:
        eeg_features[f'{col}_mean'] = eeg_window[col].mean()



    facial_features = {}
    facial_cols = ['Anger', 'Contempt', 'Disgust', 'Fear', 'Joy', 'Sadness',
       'Surprise', 'Engagement', 'Valence', 'Sentimentality', 'Confusion',
       'Neutral', 'Attention', 'Brow Furrow', 'Brow Raise', 'Cheek Raise',
       'Chin Raise', 'Dimpler', 'Eye Closure', 'Eye Widen', 'Inner Brow Raise',
       'Jaw Drop', 'Lip Corner Depressor', 'Lip Press', 'Lip Pucker',
       'Lip Stretch', 'Lip Suck', 'Lid Tighten', 'Mouth Open', 'Nose Wrinkle',
       'Smile', 'Smirk', 'Upper Lip Raise', 'Blink', 'BlinkRate', 'Pitch',
       'Yaw', 'Roll', 'Interocular Distance']

    for col in facial_cols:
        facial_features[f'{col}_mean'] = facial_window[col].mean()



    window_features = {**eeg_features, **facial_features}
    return window_features
```

```python
def get_window_label(facial_data, window_start, window_end, method='max_avg'):

    facial_window = facial_data[(facial_data['Timestamp'] >= window_start) &
                               (facial_data['Timestamp'] < window_end)]

    emotion_columns = [ 'Anger', 'Contempt', 'Disgust', 'Fear', 'Joy', 'Sadness',
       'Surprise', 'Engagement', 'Valence', 'Sentimentality', 'Confusion',
       'Neutral', 'Attention']

    if method == 'max_avg':

        avg_emotions = facial_window[emotion_columns].mean()
        dominant_emotion = avg_emotions.idxmax()

    return dominant_emotion
```

```python
def create_windowed_dataset(eeg_file, facial_file, window_size=3, overlap=1):


    eeg_data = pd.read_csv(eeg_file)
    facial_data = pd.read_csv(facial_file)


    eeg_data['TimeStamp'] = pd.to_datetime(eeg_data['TimeStamp'])
    facial_data['Timestamp'] = pd.to_datetime(facial_data['Timestamp'])


    start_time = max(eeg_data['TimeStamp'].min(), facial_data['Timestamp'].min())
    end_time = min(eeg_data['TimeStamp'].max(), facial_data['Timestamp'].max())


    windows = create_time_windows(start_time, end_time,
                                window_size=pd.Timedelta(seconds=window_size),
                                overlap=pd.Timedelta(seconds=overlap))


    dataset = []
    for window_start, window_end in windows:

        features = aggregate_window_data(eeg_data, facial_data, window_start, window_end)


        label = get_window_label(facial_data, window_start, window_end)


        sample = {
            'window_start': window_start,
            'window_end': window_end,
            'label': label,
            **features
        }
        dataset.append(sample)

    return pd.DataFrame(dataset)

# Usage
windowed_df = create_windowed_dataset('/content/drive/MyDrive/STData/1/1_EEG.csv', '/content/drive/MyDrive/STData/1/1_TIVA.csv', window_size=3, overlap=1)
print(f"Created {len(windowed_df)} training samples")
print(windowed_df.columns)
```

```python
import pandas as pd
import os
from pathlib import Path
import glob


base_dir = "/content/drive/MyDrive/STData"

def combine_csv_files(base_directory, file_type):
    """
    Combine all CSV files of a specific type (EEG or TIVA) from all numbered folders

    Parameters:
    base_directory (str): Path to the STData folder
    file_type (str): Either 'EEG' or 'TIVA'

    Returns:
    pandas.DataFrame: Combined dataframe
    """

    combined_data = []
    processed_folders = []


    for folder_num in range(1, 11):
        folder_path = os.path.join(base_directory, str(folder_num))


        if os.path.exists(folder_path):

            csv_filename = f"{folder_num}_{file_type}.csv"
            csv_path = os.path.join(folder_path, csv_filename)

            if os.path.exists(csv_path):
                try:

                    df = pd.read_csv(csv_path)


                    df['source_folder'] = folder_num


                    combined_data.append(df)
                    processed_folders.append(folder_num)

                    print(f"✓ Processed {csv_filename} from folder {folder_num} - Shape: {df.shape}")

                except Exception as e:
                    print(f"✗ Error reading {csv_path}: {str(e)}")
            else:
                print(f"✗ File not found: {csv_path}")
        else:
            print(f"✗ Folder not found: {folder_path}")


    if combined_data:
        result_df = pd.concat(combined_data, ignore_index=True)
        print(f"\n📊 Combined {file_type} data summary:")
        print(f"   - Total rows: {len(result_df)}")
        print(f"   - Total columns: {len(result_df.columns)}")
        print(f"   - Folders processed: {len(processed_folders)} ({processed_folders})")
        print(f"   - Column names: {list(result_df.columns)}")

        return result_df
    else:
        print(f" No {file_type} files were successfully processed!")
        return pd.DataFrame()


print(" Starting CSV file combination process...\n")


print("=" * 50)
print("Processing EEG files...")
print("=" * 50)
combined_eeg = combine_csv_files(base_dir, "EEG")

print("\n" + "=" * 50)
print("Processing TIVA files...")
print("=" * 50)
combined_tiva = combine_csv_files(base_dir, "TIVA")


if not combined_eeg.empty:
    eeg_output_path = "/content/drive/MyDrive/combined_EEG_data(1-10).csv"
    combined_eeg.to_csv(eeg_output_path, index=False)
    print(f"\n Combined EEG data saved to: {eeg_output_path}")


    print("\n First 5 rows of combined EEG data:")
    print(combined_eeg.head())

if not combined_tiva.empty:
    tiva_output_path = "/content/drive/MyDrive/combined_TIVA_data(1-10).csv"
    combined_tiva.to_csv(tiva_output_path, index=False)
    print(f"\n Combined TIVA data saved to: {tiva_output_path}")


    print("\n First 5 rows of combined TIVA data:")
    print(combined_tiva.head())

print("\n Process completed!")

if not combined_eeg.empty:
    print(f"\n EEG Data Info:")
    print(f"   Shape: {combined_eeg.shape}")
    print(f"   Memory usage: {combined_eeg.memory_usage().sum() / 1024**2:.2f} MB")

if not combined_tiva.empty:
    print(f"\n📈TIVA Data Info:")
    print(f"   Shape: {combined_tiva.shape}")
    print(f"   Memory usage: {combined_tiva.memory_usage().sum() / 1024**2:.2f} MB")
```

```python
import pandas as pd
import numpy as np
from datetime import timedelta
from collections import Counter
from scipy import stats
from sklearn.cluster import KMeans

def create_time_windows(start_time, end_time, window_size, overlap):

    windows = []
    current_start = start_time
    step_size = window_size - overlap

    while current_start + window_size <= end_time:
        current_end = current_start + window_size
        windows.append((current_start, current_end))
        current_start += step_size

    return windows

def aggregate_window_data(eeg_data, facial_data, window_start, window_end):

    eeg_window = eeg_data[(eeg_data['TimeStamp'] >= window_start) &
                         (eeg_data['TimeStamp'] < window_end)]
    facial_window = facial_data[(facial_data['Timestamp'] >= window_start) &
                               (facial_data['Timestamp'] < window_end)]


    eeg_features = {}
    eeg_cols = ['Delta_TP9', 'Delta_AF7', 'Delta_AF8', 'Delta_TP10',
       'Theta_TP9', 'Theta_AF7', 'Theta_AF8', 'Theta_TP10', 'Alpha_TP9',
       'Alpha_AF7', 'Alpha_AF8', 'Alpha_TP10', 'Beta_TP9', 'Beta_AF7',
       'Beta_AF8', 'Beta_TP10', 'Gamma_TP9', 'Gamma_AF7', 'Gamma_AF8',
       'Gamma_TP10', 'RAW_TP9', 'RAW_AF7', 'RAW_AF8', 'RAW_TP10']

    for col in eeg_cols:
        if not eeg_window[col].empty and len(eeg_window[col]) > 0:
            eeg_features[f'{col}_mean'] = eeg_window[col].mean()
            eeg_features[f'{col}_std'] = eeg_window[col].std()
            eeg_features[f'{col}_slope'] = calculate_slope(eeg_window[col])
        else:
            eeg_features[f'{col}_mean'] = 0
            eeg_features[f'{col}_std'] = 0
            eeg_features[f'{col}_slope'] = 0

    facial_features = {}
    facial_cols = ['Anger', 'Contempt', 'Disgust', 'Fear', 'Joy', 'Sadness',
       'Surprise', 'Engagement', 'Valence', 'Sentimentality', 'Confusion',
       'Neutral', 'Attention']

    for col in facial_cols:
        if not facial_window[col].empty and len(facial_window[col]) > 0:
            facial_features[f'{col}_mean'] = facial_window[col].mean()
            facial_features[f'{col}_max'] = facial_window[col].max()
            facial_features[f'{col}_min'] = facial_window[col].min()
            facial_features[f'{col}_std'] = facial_window[col].std()
            facial_features[f'{col}_trend'] = calculate_slope(facial_window[col])
        else:
            facial_features[f'{col}_mean'] = 0
            facial_features[f'{col}_max'] = 0
            facial_features[f'{col}_min'] = 0
            facial_features[f'{col}_std'] = 0
            facial_features[f'{col}_trend'] = 0

    return {**eeg_features, **facial_features}

def calculate_slope(series):

    if len(series) < 2:
        return 0
    x = np.arange(len(series))
    slope, _, _, _, _ = stats.linregress(x, series.values)
    return slope if not np.isnan(slope) else 0


def adaptive_threshold_labeling(facial_window, emotion_columns, percentile=75):

    if facial_window.empty:
        return 'Neutral'

    thresholds = {}
    for emotion in emotion_columns:
        if emotion != 'Neutral':
            threshold = np.percentile(facial_window[emotion], percentile)
            thresholds[emotion] = max(threshold, 0.1)


    avg_emotions = facial_window[emotion_columns].mean()
    active_emotions = []

    for emotion in emotion_columns:
        if emotion != 'Neutral' and avg_emotions[emotion] > thresholds.get(emotion, 0.1):
            active_emotions.append((emotion, avg_emotions[emotion]))

    if active_emotions:

        return max(active_emotions, key=lambda x: x[1])[0]
    else:
        return 'Neutral'

def temporal_pattern_labeling(facial_window, emotion_columns, min_duration_ratio=0.2,
                              threshold=0.1, window_size_seconds=1.0):

    if facial_window.empty or len(facial_window) < 1:
        return 'Neutral'

    results = []


    if window_size_seconds <= 1.0:
        min_duration_ratio = 0.15
        threshold = 0.08

    for emotion in emotion_columns:
        if emotion == 'Neutral':
            continue

        emotion_series = facial_window[emotion]


        if len(emotion_series) <= 3:
            avg_intensity = emotion_series.mean()
            peak_intensity = emotion_series.max()
            if avg_intensity > threshold or peak_intensity > threshold * 2:
                results.append((emotion, avg_intensity + peak_intensity))
        else:

            above_threshold = emotion_series > threshold
            sustained_duration = above_threshold.sum() / len(emotion_series)

            if sustained_duration >= min_duration_ratio:
                avg_intensity = emotion_series.mean()
                peak_intensity = emotion_series.max()
                results.append((emotion, avg_intensity * peak_intensity * sustained_duration))

    if results:
        return max(results, key=lambda x: x[1])[0]
    else:
        return 'Neutral'

def intensity_change_labeling(facial_window, emotion_columns, min_change=0.15):

    if facial_window.empty or len(facial_window) < 3:
        return 'Neutral'

    candidates = []

    for emotion in emotion_columns:
        if emotion == 'Neutral':
            continue

        emotion_series = facial_window[emotion]


        start_avg = emotion_series.head(len(emotion_series)//3).mean()
        end_avg = emotion_series.tail(len(emotion_series)//3).mean()
        intensity_change = end_avg - start_avg


        overall_intensity = emotion_series.mean()

        if intensity_change > min_change or overall_intensity > 0.3:
            score = intensity_change + overall_intensity
            candidates.append((emotion, score))

    if candidates:
        return max(candidates, key=lambda x: x[1])[0]
    else:
        return 'Neutral'

def multi_criteria_labeling(facial_window, emotion_columns):

    if facial_window.empty:
        return 'Neutral'

    emotion_scores = {}

    for emotion in emotion_columns:
        if emotion == 'Neutral':
            continue

        emotion_series = facial_window[emotion]

        avg_intensity = emotion_series.mean()


        peak_intensity = emotion_series.max()


        consistency = 1 / (1 + emotion_series.std() / (avg_intensity + 0.001))


        trend = calculate_slope(emotion_series)
        trend_score = max(0, trend)


        duration_score = (emotion_series > 0.15).sum() / len(emotion_series)

        combined_score = (
            avg_intensity * 0.3 +
            peak_intensity * 0.2 +
            consistency * 0.2 +
            trend_score * 0.15 +
            duration_score * 0.15
        )

        emotion_scores[emotion] = combined_score


    if emotion_scores:
        best_emotion = max(emotion_scores, key=emotion_scores.get)
        best_score = emotion_scores[best_emotion]

        if best_score > 0.25:
            return best_emotion

    return 'Neutral'

def context_aware_labeling(facial_window, emotion_columns, context_window_size=5):

    if facial_window.empty:
        return 'Neutral'

    emotion_candidates = []

    for emotion in emotion_columns:
        if emotion == 'Neutral':
            continue

        emotion_series = facial_window[emotion]


        if len(emotion_series) >= context_window_size:
            rolling_mean = emotion_series.rolling(context_window_size).mean()
            rolling_std = emotion_series.rolling(context_window_size).std()

            peaks = emotion_series > (rolling_mean + rolling_std)
            peak_ratio = peaks.sum() / len(emotion_series)

            score = emotion_series.mean() + peak_ratio * 0.5
            emotion_candidates.append((emotion, score))

    if emotion_candidates:
        best_emotion = max(emotion_candidates, key=lambda x: x[1])
        if best_emotion[1] > 0.2:
            return best_emotion[0]

    return 'Neutral'

def ensemble_labeling(facial_window, emotion_columns):

    if facial_window.empty:
        return 'Neutral'

    methods = [
        adaptive_threshold_labeling(facial_window, emotion_columns),
        temporal_pattern_labeling(facial_window, emotion_columns),
        intensity_change_labeling(facial_window, emotion_columns),
        multi_criteria_labeling(facial_window, emotion_columns),
        context_aware_labeling(facial_window, emotion_columns)
    ]


    votes = Counter(methods)

    most_common = votes.most_common(1)[0]
    if most_common[1] >= 3 and most_common[0] != 'Neutral':
        return most_common[0]


    return multi_criteria_labeling(facial_window, emotion_columns)


def get_advanced_window_label(facial_data, window_start, window_end, method='temporal_pattern',
                             window_size_seconds=1.0):

    facial_window = facial_data[(facial_data['Timestamp'] >= window_start) &
                               (facial_data['Timestamp'] < window_end)]

    emotion_columns = ['Anger', 'Contempt', 'Disgust', 'Fear', 'Joy', 'Sadness',
                      'Surprise', 'Engagement', 'Valence', 'Sentimentality',
                      'Confusion', 'Neutral', 'Attention']

    if method == 'adaptive_threshold':
        return adaptive_threshold_labeling(facial_window, emotion_columns)
    elif method == 'temporal_pattern':
        return temporal_pattern_labeling(facial_window, emotion_columns,
                                       window_size_seconds=window_size_seconds)
    elif method == 'intensity_change':
        return intensity_change_labeling(facial_window, emotion_columns)
    elif method == 'multi_criteria':
        return multi_criteria_labeling(facial_window, emotion_columns)
    elif method == 'context_aware':
        return context_aware_labeling(facial_window, emotion_columns)
    elif method == 'ensemble':
        return ensemble_labeling(facial_window, emotion_columns)
    else:
        return temporal_pattern_labeling(facial_window, emotion_columns,
                                       window_size_seconds=window_size_seconds)

def create_optimized_windowed_dataset(eeg_file, facial_file, labeling_method='temporal_pattern',
                                     use_short_windows=True):

    eeg_data = pd.read_csv(eeg_file)
    facial_data = pd.read_csv(facial_file)


    eeg_data['TimeStamp'] = pd.to_datetime(eeg_data['TimeStamp'])
    facial_data['Timestamp'] = pd.to_datetime(facial_data['Timestamp'])

    start_time = max(eeg_data['TimeStamp'].min(), facial_data['Timestamp'].min())
    end_time = min(eeg_data['TimeStamp'].max(), facial_data['Timestamp'].max())

    all_samples = []

    if use_short_windows:

        window_configs = [
            (0.5, 0.1),
            (1.0, 0.2),
            (1.0, 0.5),
            (1.5, 0.3),
        ]
    else:

        window_configs = [
            (1.5, 0.2),
            (2.0, 0.5),
            (3.0, 1.0),
            (4.0, 1.5),
        ]

    for window_size, overlap in window_configs:
        windows = create_time_windows(start_time, end_time,
                                    window_size=pd.Timedelta(seconds=window_size),
                                    overlap=pd.Timedelta(seconds=overlap))

        for window_start, window_end in windows:

            features = aggregate_window_data(eeg_data, facial_data, window_start, window_end)


            label = get_advanced_window_label(facial_data, window_start, window_end,
                                            method=labeling_method, window_size_seconds=window_size)

            sample = {
                'window_start': window_start,
                'window_end': window_end,
                'window_size': window_size,
                'label': label,
                **features
            }
            all_samples.append(sample)

    df = pd.DataFrame(all_samples)


    feature_columns = [col for col in df.columns
                      if col not in ['window_start', 'window_end', 'window_size', 'label']]


    df = df.drop_duplicates(subset=feature_columns + ['label'])

    return df

def analyze_and_optimize_labeling(eeg_file, facial_file, use_short_windows=True):

    methods = [
        'temporal_pattern',
        'adaptive_threshold',
        'intensity_change',
        'multi_criteria',
        'ensemble'
    ]

    results = {}

    for method in methods:
        print(f"\nTesting {method} labeling with 1-second windows...")
        df = create_optimized_windowed_dataset(eeg_file, facial_file, method, use_short_windows)

        label_counts = df['label'].value_counts()
        print(f"Distribution: {dict(label_counts)}")


        non_neutral_count = sum(count for label, count in label_counts.items() if label != 'Neutral')
        total_count = len(df)
        diversity_score = non_neutral_count / total_count if total_count > 0 else 0
        num_classes = len(label_counts)

        if num_classes > 1:
            class_proportions = np.array(list(label_counts.values)) / total_count
            balance_score = 1 - np.std(class_proportions)
        else:
            balance_score = 0

        results[method] = {
            'dataframe': df,
            'distribution': dict(label_counts),
            'diversity_score': diversity_score,
            'balance_score': balance_score,
            'total_samples': total_count,
            'non_neutral_samples': non_neutral_count,
            'num_classes': num_classes
        }


    best_method = max(results.keys(),
                     key=lambda x: results[x]['num_classes'] * results[x]['diversity_score'])

    print(f"\n=== BEST METHOD: {best_method} ===")
    print(f"Classes detected: {results[best_method]['num_classes']}")
    print(f"Diversity Score: {results[best_method]['diversity_score']:.3f}")
    print(f"Total samples: {results[best_method]['total_samples']}")
    print(f"Distribution: {results[best_method]['distribution']}")

    return results[best_method]['dataframe'], results


def create_micro_expression_dataset(eeg_file, facial_file):

    eeg_data = pd.read_csv(eeg_file)
    facial_data = pd.read_csv(facial_file)


    eeg_data['TimeStamp'] = pd.to_datetime(eeg_data['TimeStamp'])
    facial_data['Timestamp'] = pd.to_datetime(facial_data['Timestamp'])


    start_time = max(eeg_data['TimeStamp'].min(), facial_data['Timestamp'].min())
    end_time = min(eeg_data['TimeStamp'].max(), facial_data['Timestamp'].max())


    micro_windows = [
        (0.3, 0.05),
        (0.5, 0.1),
        (0.8, 0.15),
        (1.0, 0.2),
    ]

    all_samples = []

    for window_size, overlap in micro_windows:
        windows = create_time_windows(start_time, end_time,
                                    window_size=pd.Timedelta(seconds=window_size),
                                    overlap=pd.Timedelta(seconds=overlap))

        for window_start, window_end in windows:
            facial_window = facial_data[(facial_data['Timestamp'] >= window_start) &
                                       (facial_data['Timestamp'] < window_end)]

            if facial_window.empty:
                continue

            # Very sensitive labeling for micro-expressions
            emotion_columns = ['Anger', 'Contempt', 'Disgust', 'Fear', 'Joy', 'Sadness',
                             'Surprise', 'Engagement', 'Valence', 'Sentimentality',
                             'Confusion', 'Neutral', 'Attention']

            # Find ANY emotion above very low threshold
            max_emotions = facial_window[emotion_columns].max()
            label = 'Neutral'

            for emotion in emotion_columns:
                if emotion != 'Neutral' and max_emotions[emotion] > 0.05:  # Very low threshold
                    if label == 'Neutral' or max_emotions[emotion] > max_emotions[label]:
                        label = emotion

            # Aggregate features
            features = aggregate_window_data(eeg_data, facial_data, window_start, window_end)

            sample = {
                'window_start': window_start,
                'window_end': window_end,
                'window_size': window_size,
                'label': label,
                'max_emotion_intensity': max_emotions[label] if label != 'Neutral' else 0,
                **features
            }
            all_samples.append(sample)

    df = pd.DataFrame(all_samples)

    # Remove low-quality samples (very low intensities)
    df = df[df['max_emotion_intensity'] > 0.03]

    return df

def emotion_grouping_strategy(df):
    """
    Group similar emotions to create more balanced classes
    """
    def group_emotions(label):
        positive_emotions = ['Joy', 'Engagement', 'Valence']
        negative_emotions = ['Anger', 'Sadness', 'Fear', 'Disgust', 'Contempt']
        cognitive_emotions = ['Attention', 'Confusion']
        social_emotions = ['Surprise', 'Sentimentality']

        if label in positive_emotions:
            return 'Positive'
        elif label in negative_emotions:
            return 'Negative'
        elif label in cognitive_emotions:
            return 'Cognitive'
        elif label in social_emotions:
            return 'Social'
        else:
            return 'Neutral'

    df['grouped_label'] = df['label'].apply(group_emotions)

    print("Grouped emotion distribution:")
    print(df['grouped_label'].value_counts())

    return df
```
